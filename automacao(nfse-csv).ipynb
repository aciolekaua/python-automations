Automações NFSE, CSV(com prints)

```python
from selenium import webdriver
import time
import pytest
import pyautogui
import pyscreenshot
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select


#INICIO DA AUTOMAÇÃO
navegador = webdriver.Chrome()
navegador.maximize_window()
#LOGANDO NO SITE
navegador.get("https://nfse.recife.pe.gov.br/senhaweb/login.aspx")
time.sleep(3)
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[1]/div[1]/div/div[1]/input[1]").send_keys("00.000.000/0000-00")
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[1]/div[1]/div/div[1]/input[2]").send_keys("senha")

#INCIO DA CONSULTA
time.sleep(15)
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[1]/div[2]/input").click()
time.sleep(15)
pyautogui.moveTo(x=358, y=432)
time.sleep(5)
pyautogui.click(x=535, y=433)
time.sleep(5)
#SELECIONANDO EMPRESA
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/fieldset/div[1]/select").click()
time.sleep(5)
numEmpre = 22
pyautogui.press('up', presses=numEmpre)
time.sleep(14)
pyautogui.press('enter')
time.sleep(4)
#FILTRO DE DATAS
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[3]/div[1]/fieldset/div[1]/fieldset/div[1]/select").click()
time.sleep(2)
pyautogui.press('down', presses=2)
time.sleep(2)
pyautogui.press('enter')
time.sleep(5)
#INSERIR DATAS
#INSERINDO AS DATAS DE FORMA MANUAL POR ENQUANTO

#Gerando Imagens(para saber a quantidade de notas emitidas)
time.sleep(10)
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[3]/div[1]/div/input").click()
time.sleep(5)
navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[4]/div/div/fieldset/div[2]/fieldset/div/input").click()
time.sleep(5)
image = pyscreenshot.grab()
#image.show()
time.sleep(5)
nomeEmpre = "NOME DA IMAGEM"
image.save(nomeEmpre+"notastotais.png")
time.sleep(5)

navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[4]/div/div/div[1]/input").click()

#Nomes de Todas as empresas
nomes = ["NOMES DAS EMPRESAS"]

nomeEmpre_armazenado = None

#Looping que baixa todas as cv´s e coloca nome nas imagens
for nome in nomes:
    nomeEmpre_armazenado = nome
    navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/fieldset/div[1]/select").click()
    time.sleep(5)
    numEmpre = 1
    pyautogui.press('down', presses=numEmpre)
    time.sleep(5)
    pyautogui.press('enter')
    time.sleep(5)
    navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[5]/div/div/div[3]/div[3]/div[1]/div/input").click()
    time.sleep(5)
    navegador.find_element("xpath","/html/body/form/div[3]/div[1]/div[4]/div/div/fieldset/div[2]/fieldset/div/input").click()
    time.sleep(5)
    image = pyscreenshot.grab()
    #image.show()
    time.sleep(5)
    nomeEmpre = nomeEmpre_armazenado
    image.save(nomeEmpre+"notastotais.png")
    time.sleep(5)
navegador.quit()
```
